```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
|df = pd.read_csv("dm_office_sales.csv")
df.head()
```

```python
df.info()
```

```python
#scatterplot
plt.figure(figsize = (12,8))
sns.scatterplot(x='work experience',y='sales',data=df)
plt.title('this is scatter plot')
```

```python
sns.scatterplot(x = 'salary', y = 'sales', data=df, hue = 'level of education', style = 'division')
plt.legend(loc = (1.1,0.5))
```

```python
#displot
sns.displot(data=df,x='salary')
```

```python
#histogram
sns.set(style='darkgrid')
sns.histplot(data=df,x='sales', bins = 10)
```

```python
#KDE is kernel denisty estimation maps an estimate of probability density function of a random variable.
# KDE is fundamental smoothing problem.
```

```python
sns.histplot(data=df,x='sales', bins = 100, kde= True)
```

```python
import numpy as np
sample = np.random.randint(0,80,200)
sample_age_df = pd.DataFrame(sample,columns=['age'])
sns.histplot(data=sample_age_df,x='age',kde=True, bins=100)
```

```python
sns.kdeplot(data=sample_age_df,x='age')
```

```python
#countplot
plt.figure(figsize = (12,8))
sns.countplot(x='division',data=df,hue= 'level of education')
```

```python
#barplot()
sns.barplot(x='division', y='sales',data=df, estimator = np.min)
```

```python
#pairplot

sns.pairplot(df, hue = 'division', diag_kind='hist', corner = True)
```
