```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('mouse_viral_study.csv')
df.head()
```

```python
sns.scatterplot(data=df,x='Med_1_mL',y='Med_2_mL',hue='Virus Present')

x = np.linspace(0,10,150)
m=-1
b = 11
y = m*x + b
plt.plot(x,y,'g')
```

```python
from sklearn.svm import SVC
```

```python
#help(SVC)
```

```python
y = df['Virus Present']
X = df.drop('Virus Present',axis =1)
```

```python
model = SVC(kernel='linear',C = 1000)
```

```python
model.fit(X,y)
```

```python
from svm_margin_plot import plot_svm_boundary
```

```python
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='linear',C = 0.05)
model.fit(X,y)
```

```python
plot_svm_boundary(model,X,y)
```

## RBF

```python
model = SVC(kernel='rbf',C=.5)
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

## Poly

```python
model = SVC(kernel='poly',degree=19,C=.0001)
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf',C=1, gamma = 0.5)
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
#help(SVC)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
svm = SVC()
param_grid = {'C':[0.001,0.01,0.1,0.5,1], 'kernel':['linear','rbf']}
grid = GridSearchCV(svm,param_grid)
grid.fit(X,y)
```
