```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10)
y = 2*x
z = x**2
```

```python
x
```

```python
y
```

```python
#line plot
plt.plot(x,y)
plt.xlabel('Put X axis label here')
plt.ylabel('Put Y axis label here')
plt.xlim(0,6)
plt.ylim(0,10)
plt.title('Put title of the plot here')

plt.savefig('example.png')
plt.show()
```

### OOPS approach

```python
#create a figure (empty canvas)

fig = plt.figure()

#Add set of axes

axes = fig.add_axes([0,0,1,1])  #Left,bottom,width,height

#plot on that set of axes
axes.plot(x,y)

plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.3,0.5,0.5])
axes1.legend()
axes2.plot(x,y)
axes1.plot(x,z)
plt.savefig('example2.png')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x_square')
ax.plot(x,x**3, label='x_cube')
ax.legend(loc = 4)

#loc=1: upper right corner
#2: upper left corner
#3: lower left corner
#4: lower right corner
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x_square')
ax.plot(x,x**3, label='x_cube')
ax.legend(loc = (1.1,0.5))
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x_square', color = 'blue', linestyle = '-.', linewidth = 4)
ax.plot(x,x**3, label='x_cube',color='green',linestyle='--', linewidth = 6)
ax.legend(loc = 4)
```
