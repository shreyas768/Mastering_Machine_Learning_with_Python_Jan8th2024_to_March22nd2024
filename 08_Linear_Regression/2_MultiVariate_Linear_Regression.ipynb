```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Advertising.csv')
df.head()
```

```python
df.shape
```

```python
fig, axes = plt.subplots(nrows=1,ncols=3, figsize=(16,6))

axes[0].plot(df['TV'],df['sales'],'o' )
axes[0].set_title('sales vs TV spend')

axes[1].plot(df['radio'],df['sales'],'o')
axes[1].set_title('sales vs radio')

axes[2].plot(df['newspaper'],df['sales'],'o')
axes[2].set_title('sales vs newspaper')
```

```python
sns.pairplot(df)
```

## Creating features and y label

```python
X = df.drop('sales',axis=1)
y = df['sales']
```

```python
from sklearn.model_selection import train_test_split
X_train ,X_test, y_train , y_test = train_test_split(X,y,test_size= 1/3,random_state=100)
```

```python
X_train
```

```python
y_train
```

```python
y_test.shape
```

## Create Model

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

# Calculate performance of Test Set

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
df['sales'].mean()
```

```python
MAE = mean_absolute_error(y_test,y_pred)
MAE
```

```python
MSE = mean_squared_error(y_test,y_pred)
MSE
```

```python
RMSE = np.sqrt(MSE)
RMSE
```

```python
sns.histplot(data=df,x='sales')
```

## Plotting Residuals

```python
test_residuals = y_test - y_pred
```

```python
sns.scatterplot(x=y_test, y=test_residuals)
plt.axhline(y=0, c='r')
```

```python
sns.displot(test_residuals,kde=True)
```

## Retraining Model on Full Data

```python
final_model = LinearRegression()
```

```python
final_model.fit(X,y)
```

```python
y_hat = final_model.predict(X)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=3, figsize=(16,6))

axes[0].plot(df['TV'],df['sales'],'o' )
axes[0].plot(df['TV'],y_hat,'o' , color='red')
axes[0].set_title('sales vs TV spend')

axes[1].plot(df['radio'],df['sales'],'o')
axes[1].plot(df['radio'],y_hat,'o', color='red')
axes[1].set_title('sales vs radio')

axes[2].plot(df['newspaper'],df['sales'],'o')
axes[2].plot(df['newspaper'],y_hat,'o', color='red')
axes[2].set_title('sales vs newspaper')
```

```python
res = y_hat-y
```

```python
sns.scatterplot(x=y,y=res)
plt.axhline(y=0, c='r')
```

1. Holding all other features fixed, a 1 unit (1000 dollars) increase in TV spend is associated with increase in 0.045 'sales units', in this case 45 units
2. Holding all other features fixed, a  1 unit (1000 dollars) increase in Radio is associated with increase in 0.188 'sales units', in this  case 188 units
3. Holding all other features fixed, a 1 unit (1000 dollars) increase in newspaper is associated with decrease in 0.001 'sales units', in this case 1 units
